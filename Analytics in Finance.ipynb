Dependent Variable

The definition of 'good' or 'bad' cleint is not given and should be created. For the purpose of creating a model that predicts whether a customer's profile would lead to a credit card approval, the payment history of each customer was categorized into three clusters: "Not Enough Data" (cluster -1), "Bad Payer" (cluster 1), and "Good Payer" (cluster 0).

Analysing Credit Decision

Customers falling into the "Not Enough Data" category (cluster -1) were those who hadn't utilized their credit card at all, indicating a lack of transaction history. A total of 4,536 customers fit this classification and were excluded from the model's construction.

Customers deemed "Bad Payers" (cluster 1) had delayed payments of over 30 days at least once in their payment history. This included customers with specific codes (1, 2, 3, 4, or 5) in their monthly history. A total of 5,350 customers were placed in this cluster, representing a higher risk profile for the bank.

Lastly, customers with a consistent record of timely payments (status "C") or payments up to 29 days late (status "0") were categorized as low risk and considered suitable profiles for credit card approval (cluster 0). The majority of observations (36,099 customers) fell into this category.

Based on these interpretations, a new dataset called "credit_decision" was created with two columns: the unique customer IDs with credit card history and their respective classification decisions (-1, 0, or 1).

```python
credit_decision = pd.DataFrame()
credit_decision['ID'] = credit_record['ID'].unique()
```

```python
def get_credit_decision(id):
    #Determines, for a given client ID, if the client is good or bad based on payment history and card usage."""
    
    # Define status constants
    GOOD_STATUS = ['0']
    BAD_STATUS = ['1', '2', '3', '4', '5']
    UNUSED_STATUS = ['X']
    
    # Filter the credit record for the given ID
    client_status = credit_record[credit_record['ID'] == id]['STATUS'].tolist()
    
    # Check if there are any bad statuses
    if any(status in BAD_STATUS for status in client_status):
        return 1
    
    # Check if all statuses are unused
    elif all(status in UNUSED_STATUS for status in client_status):
        return -1
    
    # Assume the client is a good payer
    else:
        return 0
```

```python
# Creates a 'Decision' column in credit_decision dataset and calls get_credit_decision() to get the correct cluster
credit_decision['Decision'] = credit_decision['ID'].map(get_credit_decision)

credit_decision.info()
```

Independent Variables 
The features of this dataset are subdivided into three types: continuous numeric variables, numeric categorical and non-numerical categorical variables. The first two types of variables did not need special treatment, as they were ready to be interpreted by the model. However, the non- numerical categorical variables were required to be transformed into numerical ones.

Categorical Features 
Factorizing categorical variables allows the Random Forest Classifier to assemble decision trees based on numerical rules (larger or smaller). Such a model can deal well with numeric categorical variables without needing them to be transformed into dummy variables.

```python
# Separates the categorical variables
categorical_columns = ['FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE']
categorical_data = credit_data[categorical_columns]

# Factorizes categorical variables transforming them into numerical data
numerical_categorical_data = categorical_data.apply(lambda x: pd.factorize(x)[0])
numerical_categorical_data = pd.DataFrame(numerical_categorical_data)

numerical_categorical_data.head()
```

Each category was associated with an integer number. Thus, for example, the feature CODE_GENDER, which previously contained "M" for males and "F" for females, now has 0 and 1, respectively

```python
# Updates the values of categorical variables into credit_data data frame
credit_data[categorical_columns] = numerical_categorical_data
credit_data.head()
```

Building the Random Forest Model 
The data for building the model were randomly divided into a training group with 80% of the data and a test group with the rest 20%.

The RFC Model was configured to have 1000 estimators (decision trees) with up to 4 features.

```python
y = credit_data['Decision']
X = credit_data.drop(['Decision'], axis=1)
```

```python
unbal_X_train, X_test, unbal_y_train, y_test = split(X, y, test_size=0.2, random_state=0)
```

```python
rus = RandomUnderSampler(sampling_strategy=(3/7), random_state=0)
rus_X_train, rus_y_train = rus.fit_resample(unbal_X_train, unbal_y_train)

fig, ax = plt.subplots()
client = ['Bad Clients','Good Clients']
proportions = rus_y_train.value_counts(normalize=True)
bar_colors = ['tab:green', 'tab:red']
ax.bar(client, proportions, color=bar_colors)
ax.set_ylabel('Percentage of Dataset')
ax.set_title('Resampled Data')
ax.text(1-0.1, proportions[1]/2, '{:.2%}'.format(proportions[1]), size=10)
ax.text(0-0.1, proportions[0]/2, '{:.2%}'.format(proportions[0]), size=10)

fig.show()
```

```python
unbal_rf = RFC( n_estimators = 1000, max_features = 8, random_state=0)
unbal_rf.fit(unbal_X_train, unbal_y_train)
```

```python
rus_rf = RFC( n_estimators = 1000, max_features = 8, random_state=0)
rus_rf.fit(rus_X_train, rus_y_train)
```

```python
y_pred = unbal_rf.predict(X_test)
conf_matrix = metrics.confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(conf_matrix, cmap=plt.cm.Reds, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
plt.xlabel('Predictions', fontsize=14)
plt.ylabel('Actuals', fontsize=14)
plt.title('Confusion Matrix', fontsize=14)
plt.show()

print('Accuracy:', '{:.2%}'.format(metrics.accuracy_score(y_test, y_pred)))
print('Precision:', '{:.2%}'.format(metrics.precision_score(y_test, y_pred)))
print('Recall:', '{:.2%}'.format(metrics.recall_score(y_test, y_pred)))
print('Specificity:', '{:.2%}'.format((conf_matrix[0][0])/(conf_matrix[0][0]+conf_matrix[0][1])))
print('F1:', '{:.2%}'.format(metrics.f1_score(y_test, y_pred)))
```

Feature Importance

According to Mean Decrease in Impurity (MDI) analysis, the DAYS_BIRTH feature has the most importance (25.12%), followed by DAYS_EMPLOYED (18.44%), AMT_INCOME_TOTAL (15.99%) and OCUPPATION_TYPE (8.19%). All other features have an MDI of less than 5%. The FLAG_MOBIL feature has special irrelevance in the model's prediction capacity since every customer has a registered cell phone, making its importance equal to 0%

```python
std = np.std([tree.feature_importances_ for tree in unbal_rf.estimators_], axis=0)
importances = pd.Series(unbal_rf.feature_importances_, index=X.columns).sort_values(ascending=False)

fig, ax = plt.subplots()
importances.plot.bar(yerr=std, ax=ax, color='tab:red')
ax.set_title("Feature Importances: Unbalanced Model", size=18)
ax.set_ylabel("MDI")
for i in range(len(X.columns)):
  ax.text(i+0.2, importances[i]+0.005, '{:.2%}'.format(importances[i]), size=10)
fig.set_size_inches(15, 7)
fig.set_dpi(150)
```

Conclusion 
The model did a good job overall. It was accurate most of the time, getting things right around 87.68% of the cases. It was particularly good at saying "yes" to the right people, with about 90.31% of its "yes" answers being correct. It also caught many of the real "yes" cases, about 96.12%. However, it struggled a bit in saying "no" correctly, with only 32.15% of its "no" answers being right. The overall score, called F1, was 93.12%, showing it did well in balancing everything.

One thing to remember is that the dataset was not balanced since 87% of clients were classified as good payers while only 13% were considered bad profile. Also, the dataset only had people who actually got the card, not those who were denied at first for other reasons, like unpaid debts. This might affect the model's performance in real situations. In the future, it would be a good idea to improve the model's ability to say "no" more accurately and consider these imbalances while making the model better.
